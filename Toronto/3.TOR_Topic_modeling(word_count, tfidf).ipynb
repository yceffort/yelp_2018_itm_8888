```python
import graphlab as gl

#데이터 불러오기
data = gl.SFrame('./Toronto/reviews')
data
```

# 워드 카운트 정리

```python
#Review 데이터 워드 카운트 하기
data['word_count'] = gl.text_analytics.count_words(data['text'])
data['word_count']
```

```python
#Review 데이터 Stopword 제거하기
data['trim_stop_WC'] = data['word_count'].dict_trim_by_keys(gl.text_analytics.stopwords(), True)
data['trim_stop_WC']
```

```python
#Review 데이터 Stopword 외에 추가 Word들 확인 후 제거하기
data['adv_trim'] = data['trim_stop_WC'].dict_trim_by_keys(["[][]", ",", "(+)", "+", ".","-","\\","*","/","&","da","--",
                                                  "rex","|","de","thy","~","la","!!","limitword","...","..","????,",
                                                   "=","pok챕","you'll","횪","c:","a:","@","//","it's","6,","..",
                                                          "12","[","(","....","le","1","2","je","!","w/","**","10","2",
                                                          "5","????.","m.:","g:",":::","+[","]","???????.","????????????.",
                                                           "????????.","(++)","...du","","::","6.5","??????.",
                                                           "===============","-:::-","~~","-1","k","[+]","rv","----",
                                                           "________________________________________","_","me:",
                                                           "-------------------------------------------------","xxx,",
                                                           "jff","():","das","dee","~the","#","'n'","????????,","?",
                                                           "+1:",".an","???,","(l)","dr","dr.",
                                                           "-----------------------------------------------------------------------",
                                                           "i've", "i'm", "don't", "say" , "i","didn't","wasn't","didn't","it.",
                                                           "und","der"
                                                           ], exclude=True)
```

```python
#Review 데이터 Stopword 제거 후 순서대로 보여주기
d_wc = data[['adv_trim']].stack('adv_trim', new_column_name=['word','count']).sort('count', ascending=False)
d_wc
```

```python
#그래프로 값 이 화면에 보여주기
gl.canvas.set_target('ipynb')

#차트로 보여주기
d_wc.show()
```

# TFIDF정리하기

```python
#Review 데이터 Stopword 제거 후 tfidf 진행하기
data['tfidf'] = gl.text_analytics.tf_idf(data['adv_trim'])
data['tfidf']
```

```python
#tfidf 진행 후 값이 높은 순서대로 보여주기
df_tfidf = data[['tfidf']].stack('tfidf',new_column_name=['word','tfidf']).sort('tfidf',ascending=False)
df_tfidf
```

```python
#차트로 보여주기
df_tfidf.show()
```

# Topic Modeling (Word Count, TFIDF)

```python
#Topic 모델링 설정하기 (Word Count)
topic_wc = data['adv_trim']
```

```python
#Topic 모델링 진행하기 (TF-IDF)
topic_ti = data['tfidf']
```

```python
#Topic 선정하고 iteration 5번 돌리기 (Word Count)
model_wc = gl.topic_model.create(topic_wc, num_iterations=5)
```

```python
#Topic 선정하고 iteration 5번 돌리기 (TF-IDF)
model_ti = gl.topic_model.create(topic_ti,num_iterations=5)
```

```python
#각 Topic 단어들 보여주기 (Word Count)
model_wc.get_topics()
```

```python
#각 Topic 단어들 보여주기 (TF-IDF)
model_ti.get_topics()
```

```python
#각 Topic별 단어들 보여주기(Word Count)
model_wc.get_topics(output_type='topic_words')
```

```python
#각 Topic별 단어들 보여주기(TF-IDF)
model_ti.get_topics(output_type='topic_words')
```

```python
# Topic별 예측 모델 만들기(Word Count)
pred_wc = model_wc.predict(topic_wc)
pred_wc
```

```python
# Topic별 예측 모델 만들기 (TF-IDF)
pred_ti = model_ti.predict(topic_ti)
pred_ti
```

```python
#(부가)SFrame 기능과 결합하여 특정 주제와 관련된 문서를 찾기 위해 예측을 사용
# topic 0 관련 보기(Word Count)
docs_in_topic_wc_0 = topic_wc[model_wc.predict(topic_wc) == 0]
docs_in_topic_wc_0
```
