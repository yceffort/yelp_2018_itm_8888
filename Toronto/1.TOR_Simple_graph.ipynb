```python
import graphlab as gl #그래프랩 불러오기
import graphlab.aggregate as agg #aggregate 불러오기
import matplotlib.pyplot as plt #matplotlib.pyplot 불러오기
```

```python
#리뷰 데이터 불러오기
data = gl.SFrame('./Toronto/reviews')
data
```

```python
#Review 많이 된 레스토랑 찾아보기
d_bus = data.groupby(key_columns='business_id', operations ={'count':agg.COUNT()})
data_bus = d_bus.sort('count',ascending=False)
data_bus
```

```python
#그래프로 값 보여주기
gl.canvas.set_target('ipynb')
data_bus.show(view='bar chart')
```

```python
#리뷰 많이 된 거 탑10 세로로 그래프로 보여주기
x = d_bus['business_id'].sort(ascending=False).head(10)
y = d_bus['count'].sort(ascending=False).head(10)
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, y, align='center', alpha=0.5, color = 'red')
plt.xlabel("business_id")
plt.ylabel("count")
plt.title("Most reviewed restaurant")

plt.xticks(x_pos, x)

plt.show()
```

```python
#리뷰 많이 된 거 탑10 가로로 그래프로 보여주기
x = d_bus['business_id'].sort(ascending=False).head(10)
y = d_bus['count'].sort(ascending=False).head(10)
x_pos = [i for i, _ in enumerate(x)]
plt.barh(x_pos, y, align='center', alpha=0.5, color = 'red')
plt.xlabel("business_id")
plt.ylabel("count")
plt.title("Most reviewed restaurant")

plt.yticks(x_pos, x)

plt.show()
```
