## 1. 성별을 생성해서 집어넣는다.

```python
## 이름을 기반으로한 성별 판독 데이터셋을 집어넣는다.
import pandas as pd
gender_df = pd.read_csv('https://query.data.world/s/55vgsxa7cxlptm7v6xvremvbhm6sbo')
gender_sf = gl.SFrame(gender_df)
gender_sf.head()
```

```python
def set_gender(x):
    name = x['name'].title()
    find = gender_sf.filter_by(name, 'name')
    if find.num_rows() > 0:
        if find[0]['gender'] == 1:
            return 'M';
        else:
            return 'F';
    
    else:
        return '';
```

```python
user_gf['gender'] = user_gf.apply(lambda x: set_gender(x))
user_gf.head()
```

```python
unknown_gender_list = user_gf[user_gf['gender'] != '']
unknown_gender_list.head()
```

## 2. compliment_*, Friends 컬럼을 지운다. 이 컬럼은 필요없다.
